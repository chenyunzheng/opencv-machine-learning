```python
# mean square error
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 10, 100)
y_true = np.sin(x) + np.random.rand(x.size) - 0.5
y_pred = np.sin(x)
# plot
plt.style.use('ggplot')
plt.plot(x, y_pred, linewidth=4, label='model')
plt.plot(x, y_true, 'o', label='data')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='lower left')
```

```python
import math
a = math.sqrt(sum(pow((y_true - y_pred),2)) / y_true.size)
b = sum(pow((y_true - y_pred),2)) / y_true.size
a
b
```

```python
# use np, scikit-learn to calculate the mse
from sklearn import metrics as ms
np_mse = np.mean((y_true - y_pred) ** 2)
print('np_mse =', np_mse)
sk_mse = ms.mean_squared_error(y_true, y_pred)
print('sk_mse =', sk_mse)
```

```python
# variance, R-squared
# fraction of variance unexplained
fvu = np.var(y_true - y_pred) / np.var(y_true)
print('fvu =', fvu)
# fraction of variance explained
fve = 1.0 - fvu
print('fve =', fve)
# verify with scikit-learn
ms.explained_variance_score(y_true, y_pred)

# R-squared
r2 = 1.0 - np_mse / np.var(y_true)
print('r2 =', r2)
print('metrics_r2 =', ms.r2_score(y_true, y_pred))
```
